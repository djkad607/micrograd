# Computation Tests
Here we check if our Python libraries are working properly. The libraries we test here are NumPy and Matplotlib. NumPy is used for linear algebra, tensors and fast fourier transforms. Matplotlib is used for plotting curves. We will use 'dense' NumPy arrays here and plot a 'continuous' function over $\mathbb{R}$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Activation
xs is a 'continuous' interval here. Truly speaking, it's an numerical approximation of the closed bounded interval $[-10, 10]$. We define a continuous function $\sigma: \mathbb{R} \rightarrow \mathbb{R}$ such that $\forall x \in \mathbb{R}$, $\sigma(x) = \frac{1}{1 + e^{-x}}$. Clearly $\forall x \in \mathbb{R}$, $e^{-x} > 0$

$\implies \sigma(x) > 0$.

Also, $1 + e^{-x} > 1$

$\implies \frac{1}{1 + e^{-x}} < 1$.

Therefore $\sigma(\mathbb{R}) = (0, 1)$.

ys is a numerical approximation of the open bounded interval $(0, 1)$. 

```python
xs = np.arange(-10, 10, 0.01)
ys = (1 + np.exp(-xs)) ** -1
ys
```

## Plotting
We plot the curve on the closed bounded interval $[-10, 10]$ now.

```python
plt.plot(xs, ys, label="Sigmoid"); plt.grid(True); plt.legend(); plt.title("Sigmoid Curve"); plt.xlabel("Input Space"); plt.ylabel("Output Space"); plt.show()
```
